## Introducción:

```python
### Importr librerías 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Csrgar rl dataset 
df = pd.read_csv("dataset_cars.csv", header=None)

```

```python
# 3. Análisis del perfil del dataset 
#Despliega los nombres de las columnas
df.columns
#despliega los primeros cinco registros 
df.head()
```

```python
"""El dataset no cuenta con titulos en las columnas Lo tenemos que agregar """

# Crear lista con el nombre de las columnas  
encabezados = ["symboling","normalized-losses","make","fuel-type","aspiration",
"num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
#A gregar la lista como columna del dataset 
df.columns=encabezados
# Comprobar 
df.columns
df.head()
df.tail()
```

```python
#4. Obtener un panorama de la estructura del dataset 
df.shape
```

```python
# Estadisticas basicas de tendencia central
df.describe()
```

```python
#Replace the records in the dataset that have the value “?” for Nan's. For this purpose, use the Numpy command:
df.replace("?", np.nan, inplace=True)
```

```python
# Factor de conversión 
# Crear las nuevas columnas con el cálculo:
kpl_factor = 0.425144
df["city-kpl"] = df["city-mpg"] * kpl_factor
df["autopista-kpl"] = df["highway-mpg"] * kpl_factor
# Validar el resultado
print(df[["city-mpg","city-kpl","highway-mpg","autopista-kpl"]].head())
```

# Visualización 

#### Cálculo: el precio suele venir como "object" (texto) por los signos de “?”. Hay que limpiarlo.


```python
# Limpieza rápida de la columna precio
df['price'] = pd.to_numeric(df['price'], errors='coerce')
# Eliminar filas sin precio
df.dropna(subset=['price'], inplace=True)
```

#### Gráfica: Un Boxplot de Price (precio) por make (marca).

```python
# Gráfica de cajas (Boxplot) para ver distribución por marcas
plt.figure(figsize=(12, 6))
sns.boxplot(x='price', y='make', data=df)
plt.title('Distribución de Precios por Marca')
plt.show()
```

#### Gráfica: Heatmap (Mapa de calor). Esto les enseña a identificar qué variables "se mueven juntas".

```python
# Grafica de calor para ver la correlación entre variables numéricas
corr_matrix = df[['engine-size', 'horsepower', 'curb-weight', 'city-kpl', 'price']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Mapa de calor: correlación entre variables')
plt.show()
```

#### Gráfica: Scatter Plot con Línea de Regresión.

```python
# Gráfica: Scatter Plot con Línea de Regresión
plt.figure(figsize=(10, 6))
sns.regplot(x='engine-size', y='city-kpl', data=df, color='green')
plt.title('Relación: Tamaño del Motor vs. Kilómetros por Litro')
plt.xlabel('Tamaño del Motor (Engine Size)')
plt.ylabel('Eficiencia (City KPL)')
plt.show()
```

## 4. Agrupación (Binning): De Numérico a Categórico

```python
# Crear 3 grupos (bins) de precios
# Grupo 1: Desde el precio más bajo (df['price'].min()) hasta 15,000.
# Grupo 2: Desde 15,001 hasta 30,000.
# Grupo 3: Desde 30,001 hasta el precio más alto (df['price'].max())
bins = [df['price'].min(), 15000, 30000, df['price'].max()]
nombres = ['Económico', 'Estándar', 'Lujo']
df['price-category'] = pd.cut(df['price'], bins, labels=nombres, include_lowest=True)
# Visualizar cuántos hay de cada uno
sns.countplot(x='price-category', data=df, hue='price-category', palette='Set2', legend=False)
plt.title('Distribución de Autos por Categoría de Precio')
plt.show()
```
