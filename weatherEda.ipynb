## Introducción

El estudio del clima es importante porque influye en muchas actividades de la vida cotidiana, como la agricultura, el transporte y la salud. Para comprender el comportamiento climático, es necesario analizar variables como la temperatura, la humedad, la velocidad del viento y la presión atmosférica, las cuales permiten describir las condiciones de la atmósfera.

En este documento se presenta un análisis exploratorio de datos (EDA) utilizando el dataset *Weather History*, el cual contiene información histórica de distintas variables climáticas. El objetivo es limpiar, explorar y visualizar los datos utilizando herramientas como Python, pandas, matplotlib y seaborn, con el fin de identificar patrones, tendencias y relaciones entre variables como la temperatura, la sensación térmica y la humedad.

```python
#Paso 1 Importar las librerías necesarias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Paso 2 Cargar el dataset
df = pd.read_csv('weatherHistory.csv')
```

## Descripción de columnas del dataset (estructura de lista)

A continuación se presenta la descripción de cada columna del dataset Weather History:

- **Formatted Date**: Representa la fecha y hora exacta en que se registraron los datos meteorológicos, incluyendo la zona horaria.

- **Summary**: Describe brevemente las condiciones climáticas generales observadas en ese momento (por ejemplo, "Partly Cloudy", "Clear", etc.).

- **Precip Type**: Indica el tipo de precipitación presente, como lluvia ("rain") o nieve ("snow").

- **Temperature (C)**: Muestra la temperatura real medida en grados Celsius.

- **Apparent Temperature (C)**: Representa la sensación térmica percibida por el cuerpo humano, la cual puede diferir de la temperatura real debido a factores como el viento y la humedad.

- **Humidity**: Indica la humedad relativa del aire. Sus valores van de 0 a 1, donde valores más altos representan mayor humedad.

- **Wind Speed (km/h)**: Muestra la velocidad del viento en kilómetros por hora.

- **Wind Bearing (degrees)**: Indica la dirección del viento en grados, donde 0° representa el norte.

- **Visibility (km)**: Representa la distancia máxima a la que se pueden observar objetos claramente, medida en kilómetros.

- **Pressure (millibars)**: Indica la presión atmosférica medida en milibares.

- **Loud Cover**: Columna con valores nulos o irrelevantes para este análisis, por lo que fue eliminada durante la limpieza de datos.

- **Daily Summary**: Proporciona un resumen general del clima durante el día completo, pero fue eliminada para evitar redundancia en el análisis.

## 2) Limpieza y filtrado de ruido
Se eliminan columnas administrativas o con poca variabilidad (ruido) y se revisan valores nulos/duplicados.

```python
#Describe
df.describe()
```

```python
# Conversión de la columna 'Formatted Date' a tipo datetime
# UTC=True es necesario porque los datos tienen zona horaria
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
```

```python
# Establecer la fecha como índice para facilitar el análisis temporal
df.set_index('Formatted Date', inplace=True)
```

```python
print(df.head())
```

```python
# Eliminamos columnas que no usaremos para este análisis específico
# 'Loud Cover' suele ser 0 en todo el dataset de Szeged
df.drop(columns=['Loud Cover', 'Daily Summary'], inplace=True)
```

```python
# Comprobamos si hay valores nulos
print(df.isnull().sum())
```

## 3) Temperatura real vs. sensación térmica
Gráfico de líneas con una muestra de registros para que sea legible.

```python
#Tomamos una muestra (los primeros 500 registros)  para que el grafica sea legible
df_sample = df.iloc[:500]
plt.figure(figsize=(15, 6))
plt.plot(df_sample.index, df_sample['Temperature (C)'], label='Temp Real', color='blue')
plt.plot(df_sample.index, df_sample['Apparent Temperature (C)'], label='Sensación Térmica', color='red', linestyle='--')
plt.title('Temperatura Real vs Sensación Térmica (szeged)')
plt.legend()
plt.show()
```

## 4) Remuestreo (Resampling)
El dataset es horario; para observar tendencias de una década, calculamos promedios **anuales**.

```python
# Calculamos el promedio anual de todas las variables
df_yearly = df[['Temperature (C)', 'Humidity']].resample('Y').mean()
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_yearly, x=df_yearly.index, y='Temperature (C)', marker='o')
plt.title('Tendencia de la Temperatura Promedio Anual (2006-2016)')
plt.show()
```

## 5) Correlación: Temperatura vs Humedad
Validamos visualmente la relación entre temperatura y humedad con regresión. Además calculamos el coeficiente de correlación.

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Temperature (C)', y='Humidity', data=df.sample(2000), scatter_kws={'alpha':0.2}, line_kws={'color':'red'})
plt.title('Relación Inversa entre Temperatura y Humedad')
plt.show()
```

## 6) Gráfico extra: Histograma de Wind Speed (km/h)
Histograma para entender qué tan ventosa es la región y marcar promedio/mediana.

```python
# Configuración del gráfico
plt.figure(figsize=(10, 6))
# Creación del histograma con una línea de densidad (KDE)
sns.histplot(df['Wind Speed (km/h)'], bins=30, kde=True, color='teal')

# Añadir líneas de referencia estadística
plt.axvline(df['Wind Speed (km/h)'].mean(), color='red', linestyle='--', label='Promedio')
plt.axvline(df['Wind Speed (km/h)'].median(), color='orange', linestyle='-', label='Mediana')
# Estética del gráfico
plt.title('Distribución de la Velocidad del Viento en Szeged (2006-2016)')
plt.xlabel('Velocidad del Viento (km/h)')
plt.ylabel('Frecuencia (Horas registradas)')
plt.legend()
plt.show()
```

## 8) Conclusión escrita (para entregar)
A partir de los registros horarios de Szeged (2006–2016), se observa que la **temperatura** y la **sensación térmica** siguen una relación positiva, pero en ciertos periodos la sensación térmica cae por debajo de la temperatura real, lo cual es consistente con el efecto del **viento** y la interacción con la humedad. Además, la relación entre **temperatura y humedad** muestra un patrón **inverso**: cuando la temperatura aumenta, la humedad tiende a disminuir, lo que se refleja en la tendencia y en el valor de correlación calculado.
